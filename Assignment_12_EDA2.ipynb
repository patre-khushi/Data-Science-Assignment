```python
df.dropna(inplace=True)
```

```python
# Apply scaling techniques to numerical features
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Identify numerical features
numerical_features = df.select_dtypes(include=['number']).columns

#Standard Scaling
scaler = StandardScaler()
df_standard_scaled = df.copy()
df_standard_scaled[numerical_features] = scaler.fit_transform(df[numerical_features])
df_standard_scaled.head()
```

```python
# Min-Max Scaling
scaler_minmax = MinMaxScaler()
df_minmax_scaled = df.copy()
df_minmax_scaled[numerical_features] = scaler_minmax.fit_transform(df[numerical_features])
df_minmax_scaled.head()
```

Task 2: Encoding Techniques

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
# Identify categorical features
categorical_features = df.select_dtypes(include=['object']).columns
```

```python
# Apply One-Hot Encoding to categorical variables with less than 5 categories
OHE_df = pd.get_dummies(df, columns=categorical_features[df[categorical_features].nunique() < 5])
OHE_df.head()
```

```python
# Use Label Encoding for categorical variables with more than 5 categories
label_encoder = LabelEncoder()
LE_df = df.copy()
for col in categorical_features:
  if df[col].nunique() > 5:
    LE_df[col] = label_encoder.fit_transform(df[col])
LE_df.head()
```

The pros and cons of One-Hot Encoding and Label Encoding

1.One-Hot Encoding

Pros: Avoids assuming any order among categories; suitable for categorical data.

Cons: Can lead to high dimensionality and increased model complexity.

2.Label Encoding

Pros: Space-efficient and simple to implement.

Cons: Assumes an ordinal relationship that may not exist, potentially misleading the model.








Task 3: Feature Engineering

```python
# Create at least 2 new features

df['age_group'] = pd.cut(df['age'], bins=[0, 25, 50, 75, 100], labels=['0-25', '26-50', '51-75', '76-100'])
df['hours_per_week_group'] = pd.cut(df['hours_per_week'], bins=[0, 20, 40, 60, 80], labels=['0-20', '21-40', '41-60', '61-80'])

print("Data with new features:\n", df[['age', 'age_group', 'hours_per_week', 'hours_per_week_group']].head())
```

```python
# Apply a transformation to a skewed numerical feature
# Let's apply a log transformation to the capital_gain feature if it's skewed
```

```python
# Check skewness
print("Skewness of capital_gain:", df['capital_gain'].skew())
```

```python
# Apply log transformation
df['capital_gain_log'] = np.log1p(df['capital_gain'])
```

```python
import matplotlib.pyplot as plt
```

```python
# Plot to visualize the effect
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(df['capital_gain'], bins=30)
plt.title('Original Capital Gain')
plt.subplot(1, 2, 2)
plt.hist(df['capital_gain_log'], bins=30)
plt.title('Log-transformed Capital Gain')
plt.show()
```

```python
# 4.1 Use the Isolation Forest algorithm to identify and remove outliers

from sklearn.ensemble import IsolationForest
```

```python
# Use Isolation Forest
iso_forest = IsolationForest(contamination=0.01, random_state=42)
outliers = iso_forest.fit_predict(df[numerical_features])
```

```python
# Remove outliers
df_no_outliers = df[outliers == 1]
df_no_outliers.head()
```
