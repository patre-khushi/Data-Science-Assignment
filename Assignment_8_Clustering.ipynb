Task 2:Implementing Clustering Algorithms

1. K-mean Clustering

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
inertia = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8, 4))
plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
# Optimal k found from the elbow method
optimal_k = 4  # Assuming 4 from the elbow curve
```

```python
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans_labels = kmeans.fit_predict(scaled_data)
```

```python
kmeans_labels
```

```python
# Silhouette score for K-Means
kmeans_silhouette = silhouette_score(scaled_data, kmeans_labels)
print(f'Silhouette Score for K-Means: {kmeans_silhouette}')
```

2. Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch # to build dendrogram and build the plotting
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
```

```python
# Hierarchical clustering using 'ward' linkage
Z = linkage(scaled_data, method='ward')
```

```python
plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.title('Dendrogram for Hierarchical Clustering')
plt.xlabel('Sample index')
plt.ylabel('Distance')
plt.show()
```

```python
# Choosing the number of clusters from dendrogram
hierarchical_labels = fcluster(Z, t=optimal_k, criterion='maxclust')
hierarchical_labels
```

```python
# Silhouette score for Hierarchical Clustering
hierarchical_silhouette = silhouette_score(scaled_data, hierarchical_labels)
print(f'Silhouette Score for Hierarchical Clustering: {hierarchical_silhouette}')
```

3. DBSCAN Clustering

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(scaled_data)
```

```python
# Silhouette score for DBSCAN
dbscan_silhouette = silhouette_score(scaled_data, dbscan_labels)
print(f'Silhouette Score for DBSCAN: {dbscan_silhouette}')
```

Task 3:
Cluster Analysis and Interpretation


```python
# Adding cluster labels to the original data
df['KMeans_Cluster'] = kmeans_labels
df['Hierarchical_Cluster'] = hierarchical_labels
df['DBSCAN_Cluster'] = dbscan_labels

# Analyzing clusters
kmeans_clusters = df.groupby('KMeans_Cluster').mean()
hierarchical_clusters = df.groupby('Hierarchical_Cluster').mean()
dbscan_clusters = df.groupby('DBSCAN_Cluster').mean()
```

```python
print("DBSCAN Clusters Analysis:\n", dbscan_clusters)
```
