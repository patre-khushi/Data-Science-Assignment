```python
# Outlier Detection: User defined Function to calculate Upper Extreme and Lower Extreme value
def outlier_detection(data,colname):
  q1 = data[colname].quantile(0.25)
  q3 = data[colname].quantile(0.75)
  iqr = q3 - q1

  upper_extreme = q3 + (1.5 * iqr)
  lower_extreme = q1 - (1.5 * iqr)

  return lower_extreme, upper_extreme,q1,q3
```

**Task 4: Pattern Recognition and Insights**

```python
# Select the columns LB and AC
selected_columns = df[['LB', 'AC']]

# Calculate the correlation matrix for LB and AC
correlation_matrix = selected_columns.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title('Correlation Matrix for LB and AC')
plt.show()
```

The correlation matrix will provide a single value representing the correlation between LB and AC. This value can range from -1 to 1:

A value close to 1 indicates a strong positive correlation, meaning that as LB increases, AC also tends to increase.
A value close to -1 indicates a strong negative correlation, meaning that as LB increases, AC tends to decrease.
A value close to 0 indicates no significant linear relationship between LB and AC.

The correlation matrix of LB and AC is Positive Correlation (0.26), The correlation value of 0.26 indicates a moderate to strong positive relationship between baseline FHR and accelerations.
